### Zadanie 3.1: DataFrame

```python
# TODO: Utwórz DataFrame:
# - Imie (5 imion)
# - Wiek (20-30)
# - Miasto (5 miast)
# - Ocena (3.0-5.0)
data = {
    "name": ["Alek", "Marek", "Kacper", "Jurek", "Wiktor"],
    "age": np.random.randint(20,30, size=5),
    "city": ["Poznan", "Warszawa", "Wroclaw", "Krakow", "Gdansk"],
    "grade": np.round(np.random.uniform(3.0, 5.0, 5), 1)
}

df = pd.DataFrame(data)
print(df)
```

```python
# b) Scatter: petal length vs width (kolor = gatunek)
from sklearn.datasets import load_iris

iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df['species'] = iris.target
df['species_name'] = df['species'].map({0: 'setosa', 1: 'versicolor', 2: 'virginica'})

# b) Scatter: petal length vs width (kolor = gatunek)
plt.figure(figsize=(10, 6))

# Scatter plot dla każdego gatunku
for species_id, species_name in enumerate(['setosa', 'versicolor', 'virginica']):
    mask = df['species'] == species_id
    print(mask)
    plt.scatter(df[mask]['petal length (cm)'], 
                df[mask]['petal width (cm)'],
                label=species_name,
                alpha=0.7)

plt.xlabel('Petal Length (cm)')
plt.ylabel('Petal Width (cm)')
plt.title('Iris: Petal Length vs Petal Width by Species')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

```python

# c) Histogram długości płatków
iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df['species'] = iris.target

# Histogram dla każdego gatunku
plt.figure(figsize=(10, 6))

for species_id, species_name in enumerate(['setosa', 'versicolor', 'virginica']):
    mask = df['species'] == species_id
    plt.hist(df[mask]['petal length (cm)'], 
             bins=15, 
             alpha=0.6, 
             label=species_name,
             edgecolor='black')

plt.xlabel('Petal Length (cm)')
plt.ylabel('Frequency')
plt.title('Histogram of Petal Length by Species')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

## Część 5: Mini-projekt Wine Quality

```python
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-red.csv'
wine_df = pd.read_csv(url, sep=';')
```

```python
# TODO 2: Analiza 'quality'
# - unikalne wartości, rozkład, średnia
# - wykres słupkowy

print(wine_df['quality'].mean())

plt.figure(figsize=(8, 5))

wine_df['quality'].value_counts().sort_index().plot(kind='bar')
plt.title('Rozkład ocen jakości wina')
plt.xlabel('Jakość')
plt.ylabel('Liczba win')
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# TODO 3: Korelacje
corr_matrix = wine_df.corr()
print(corr_matrix)

plt.figure(figsize=(10,10))
sns.heatmap(
    corr_matrix,
    annot=True,
    cmap='coolwarm',
    
)
```

```python
# TODO 4: Rozkłady
wine_df.hist(bins=20, figsize=(16, 12), color='coral', edgecolor='black', alpha=0.7)
plt.suptitle('Histogramy - metoda pandas', fontsize=16, y=1.00)
plt.tight_layout()


def detect_outliers_iqr(data):
   Q1 = data.quantile(0.25)
   Q3 = data.quantile(0.75)
   IQR = Q3-Q1 
   lower_bound = Q1 - 1.5 * IQR
   upper_bound = Q3 + 1.5 * IQR
   outliers = (data < lower_bound) | (data > upper_bound)
   return outliers, lower_bound, upper_bound

fig, axes = plt.subplots(4, 4, figsize=(18, 14))
fig.suptitle('Histogramy z outlierami (metoda IQR) - wszystkie cechy', 
             fontsize=16, fontweight='bold')

axes = axes.flatten()

for i, col in enumerate(df.columns):
    outliers, lower, upper = detect_outliers_iqr(df[col])
    
    axes[i].hist(df[col], bins=20, color='lightblue', edgecolor='black', alpha=0.7)
    axes[i].axvline(lower, color='red', linestyle='--', linewidth=1.5, alpha=0.7)
    axes[i].axvline(upper, color='red', linestyle='--', linewidth=1.5, alpha=0.7)
    
    # Zaznacz outliery
    if outliers.sum() > 0:
        axes[i].scatter(df[col][outliers], np.zeros(outliers.sum()), 
                       color='red', s=50, marker='x', zorder=5)
    
    axes[i].set_title(f'{col}\n({outliers.sum()} outlierów)', fontsize=9, fontweight='bold')
    axes[i].set_xlabel('Wartość', fontsize=7)
    axes[i].set_ylabel('Częstość', fontsize=7)
    axes[i].grid(True, alpha=0.3)

for i in range(len(df.columns), len(axes)):
    axes[i].axis('off')
plt.show()
```

```python
# TODO 5: Kategorie jakości
# - Low (<=5), Medium (5-6), High (>6)
# - porównanie cech
# - box plots

def categorize_quality(quality):
  if quality <= 5:
    return "Low"
  elif 5 < quality <= 6:
    return "Mid"
  else:
    return "High"


wine_df['quality_grade'] = wine_df['quality'].apply(categorize_quality)
print(wine_df.head())
```
